# Import Dependencies

We begin by importing the necessary libraries.

```python
# Web scraping
import requests

# Data analysis
import pandas as pd

# Data visualization
import folium
```

# Web Scraping

Then, we use the Rest Countries API to retrieve information about each African country and their corresponding capital.

```python
# Load shapefiles from Rest Countries API
url = 'https://restcountries.com/v3.1/region/africa'
response = requests.get(url)
data = response.json()

# Create a list of capital coordinates: (name, capital, lat, long)
capitals = [(country['name']['common'], country['capital'][0], country['latlng'][0], country['latlng'][1]) for country in data]
```

# Data Visualization

Finally, we create the base map of Africa, before adding markers as popups indicating each nation's name and capital. 

```python
# Create a base map of Africa
m = folium.Map(location=[8.7832, 34.5085], zoom_start=3)

# Add markers for each nation's capital
for name, capital, lat, lon in capitals:
    tooltip = f'Country: {name}, Capital: {capital}'
    folium.Marker([lat, lon], tooltip=tooltip).add_to(m)

# Save the map as an HTML file
m.save('africa_map.html')

# Display results
m
```
