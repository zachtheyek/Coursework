# Question 1.12

From Taylor's Classical Mechanics (2005).

```python
import numpy as np
import matplotlib.pyplot as plt

# Configure 3D plotting
ax = plt.figure().add_subplot(projection='3d')

# Define constants
b = c = v_0 = omega = np.linspace(100, 100, 100)
t = np.linspace(0, 100, 100)

# Prepare arrays x, y, z
x = b * np.cos(omega * t)
y = c * np.sin(omega * t)
z = v_0 * t

# Make plot
ax.plot(x, y, z)

# Display plot
plt.show()
```

```python
ax = plt.figure().add_subplot(projection='3d')

# Vary constants
b = c = v_0 = omega = np.linspace(1, 1, 100)

# Redefine arrays x, y, z
x = b * np.cos(omega * t)
y = c * np.sin(omega * t)
z = v_0 * t

# Make and display plot
ax.plot(x, y, z)
plt.show()
```
