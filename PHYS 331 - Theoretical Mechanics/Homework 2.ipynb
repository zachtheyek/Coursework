# Question 1.26

From Taylor’s Classical Mechanics (2005).

## a.

Equations of motion in $S$:

$$\begin{equation}
  \begin{split}
    x&=0\\
    y(t)&=vt
  \end{split}
\end{equation}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
v = 5
t = np.linspace(0, 10, 10)

# Prepare arrays x, y
x = np.linspace(0, 0, 10)
y = v * t

# Make plot
plt.plot(x, y)
plt.title('Trajectory of Puck Over Time in $S$')
plt.xlabel('x (m)')
plt.ylabel('y (m)')

# Display plot
plt.show()
```

## b.

Equations of motion in $S'$:

$$\begin{equation}
  \begin{split}
    x'(t)&=-v_xt\\
    y(t)&=v_yt
  \end{split}
\end{equation}$$

```python
# Define constants
v_x, v_y = 5, 5
t = np.linspace(0, 10, 10)

# Prepare arrays x, y
x_prime = - v_x * t
y_prime = v_y * t

# Make plot
plt.plot(x_prime, y_prime)
plt.title('Trajectory of Puck Over Time in $S\'$')
plt.xlabel('x\' (m)')
plt.ylabel('y\' (m)')

# Display plot
plt.show()
```

## c.

Equations of motion in $S''$:

$$\begin{equation}
  \begin{split}
    x''(t)&=-\frac{1}{2}at^2\\
    y''(t)&=v_yt
  \end{split}
\end{equation}$$

```python
# Define constants
v_y = 5
a = 5
t = np.linspace(0, 10, 10)

# Prepare arrays x, y
x_double_prime = - 0.5 * a * t**2
y_double_prime = v_y * t

# Make plot
plt.plot(x_double_prime, y_double_prime)
plt.title('Trajectory of Puck Over Time in $S\'\'$')
plt.xlabel('x\'\' (m)')
plt.ylabel('y\'\' (m)')

# Display plot
plt.show()
```
