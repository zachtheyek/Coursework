# Import Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Create & Visualize Data

$$\begin{equation*}
  y_\text{blue}=2.1 + 5.18x
\end{equation*}$$

$$\begin{equation*}
  y_\text{green}=1.7 + 5.30x
\end{equation*}$$

```python
df = pd.DataFrame({'X': [0.00, 0.05, 0.40, 0.73, 0.26, 0.45, 0.33, 0.61, 0.71, 0.17, 0.09, 0.12],
                   'Y': [1.96, 1.62, 4.04, 5.83, 3.60, 4.72, 3.25, 5.16, 5.23, 2.62, 2.40, 2.86]})
X, y = df['X'], df['Y']
pred_1 = 2.1 + 5.18 * X
pred_2 = 1.7 + 5.30 * X

plt.scatter(X, y, c='r')
plt.plot(X, pred_1, c='b')
plt.plot(X, pred_2, c='g')
plt.show()
```

# Define MAE & MSE

$$\begin{equation*}
  \text{MAE}=\frac{1}{n}\sum_{i=1}^n |y_i - \hat{y}_i|
\end{equation*}$$

$$\begin{equation*}
  \text{MSE}=\frac{1}{n}\sum_{i=1}^n (y_i - \hat{y}_i)^2
\end{equation*}$$

```python
def MAE(labels, predictions):
  return abs(predictions - labels).mean()

def MSE(labels, predictions):
  return ((predictions - labels)**2).mean()
```

# Compute Errors

```python
print(f'MAE of blue regression: {MAE(y, pred_1)}')
print(f'MAE of green regression: {MAE(y, pred_2)}\n')

print(f'MSE of blue regression: {MSE(y, pred_1)}')
print(f'MSE of green regression: {MSE(y, pred_2)}')
```

# Conclusion

From the MAE, it seems that the blue regression is a better fit to the data, whereas from the MSE, the green regression appears to model the data more accurately. 

Since the difference between the blue and green MAE's is about 7 times the difference between the blue and green MSE's, we conclude that the blue regression is better.
