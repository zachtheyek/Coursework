```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define constants
t = np.linspace(0, 10, 21)

# Prepare arrays x, y, z
x = (2 * t**2 - 1) / (t**2 + 1)
y = (3 * t) / (t**2 + 1)

# Make plot
plt.scatter(x, y)
plt.title('Motion of particle given by z(t)')
plt.xlabel('X')
plt.ylabel('Y')

# Display plot
plt.show()
```

```python
df = pd.DataFrame({'t': t, 'X': x, 'Y': y})
df
```
